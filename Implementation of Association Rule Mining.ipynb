```python
import numpy as np
import pandas as pd
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
import matplotlib.pyplot as plt
```

```python
basket = pd.read_csv(r"C:\jupyter\arules\datasets\groceries.csv", header=None)
basket.head()
```

```python
basket = pd.get_dummies(basket)
basket.head(100)
```

```python
frequent_itemsets = apriori(basket, min_support=0.007, use_colnames=True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.8)
rules.head(15)
```

```python
rules1 = rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
rules1.head(15)
```

```python
rules2 = rules1.drop_duplicates(subset=['support'])
rules2.head(15)
```

```python
rules3 = rules2[rules2['confidence'] >=0.20]
rules3.sort_values('confidence', ascending = False)
```

```python
x = rules3.support
y = rules3.confidence
```

```python
plt.scatter(x,y)
plt.xlabel('support')
plt.ylabel('confidence')
plt.savefig('plot1.png', dpi=300, bbox_inches='tight')
```

```python
plt.hist(x)
plt.savefig('plot2.png', dpi=300, bbox_inches='tight')
```

```python
plt.hist(y)
plt.savefig('plot3.png', dpi=300, bbox_inches='tight')
```
